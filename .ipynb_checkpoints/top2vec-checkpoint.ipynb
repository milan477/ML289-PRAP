```python
def _read_single_pdf_pytesseract(path: Path):
    contents = []

    print(f"Reading {path.name} with pytesseract")
    pages = convert_from_path(path, first_page=1, last_page=2)

    data = []

    for page in pages[:1]:

        print(f'Processing page {len(data)+1} / {len(pages)}')

        df = pd.DataFrame(
                pytesseract.image_to_data(
                    image = page,
                    lang = 'eng',
                    output_type=pytesseract.Output.DICT)
                    )

        data.append(df)

        content = display_by_line(df, printQ=False)
        contents.append(content)

        image = preprocess(page).unsqueeze(0).to(device)

        text = clip.tokenize(doc_types).to(device)


        with torch.no_grad():
            image_features = model.encode_image(image)
            text_features = model.encode_text(text)

            logits_per_image, logits_per_text = model(image, text)
            probs = logits_per_image.softmax(dim=-1).cpu().numpy()

            label = doc_types[probs.argmax()]
            print("================================= Predicted label:", label)


    return pages[0], data[0], contents[0] # TODO: limited to first page for now
```
