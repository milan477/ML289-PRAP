```python
def get_image_count(file):
  doc = pymupdf.open(file)
  for pg_idx in range(len(doc)):
    page = doc[pg_idx]
    image_list = page.get_images()
    if image_list:
      print(f"Found {len(image_list)} images on page {pg_idx}")
    else:
      print("No images found on page", pg_idx)
  return len(image_list)

    #  for image_index, img in enumerate(image_list, start=1): # enumerate the image list
    #     xref = img[0]
    #     pix = pymupdf.Pixmap(doc, xref)
    #     if pix.n - pix.alpha > 3:
    #       pix = pymupdf.Pixmap(pymupdf.csRGB, pix)
```

```python
def create_df():
    files = {"record" : [], "text_len": [], "num_images": []}
    for i in range(10):
        file = f"data/record{i}.pdf"
        md_text = pymupdf4llm.to_markdown(file)
        # md_text = ""
        # if get_text_percentage(file) > 0.05:
        #     md_text = pymupdf4llm.to_markdown(file)
        files["record"].append(file)
        files["text_len"].append(len(md_text))
        files["num_images"].append(get_image_count(file))
    files_df = pd.DataFrame(files)
    files_df.head()
    return files_df
df = create_df()
df.head()
```
